```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

## Data scaling

```python
df_overview = pd.read_csv('../../data/df_overview.csv', index_col=0)
df_overview
```

```python
df_overview.sort_values('trailing_pe', ascending=False)
```

```python
df_overview[['trailing_pe']] = df_overview[['trailing_pe']].dropna().values.astype('float32')
df_overview
```

### Trailing P/E ratio

```python
# Normalize ESG-Score and market-capital
scaler = MinMaxScaler()
df_overview[['trailing_pe_scaled']] = scaler.fit_transform(df_overview[['trailing_pe']])
```
